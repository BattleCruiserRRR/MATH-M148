```python
# START PREP DATA FOR ALGO

# data manipulation to get frequencies of each event per customer/account pair

# select cols
export_head = export_head[["customer_id","account_id","ed_id"]]
# select customer/account/ed_id, count each observation
export_head["count"] = export_head.groupby(by = ["customer_id","account_id","ed_id"])["ed_id"].transform("count")
# groups are duplicated, so remove duplicates
export_head = export_head.drop_duplicates()
```
